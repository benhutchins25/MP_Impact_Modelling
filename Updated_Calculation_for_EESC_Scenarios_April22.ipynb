```python
# for each of these plots we want to perform four different fits to identify the one with the best r2 score
# first lets just try to do this with NoMP1987

# first lets perform the polyfits

# define x and y values

x_mp = df_NoMP1987['Time (year)']
y_mp = df_NoMP1987['EESC (ppt)']

x_la = df_NoLA1990['Time (year)']
y_la = df_NoLA1990['EESC (ppt)']

x_ca = df_NoCA1992['Time (year)']
y_ca = df_NoCA1992['EESC (ppt)']

x_ba = df_NoBA1999['Time (year)']
y_ba = df_NoBA1999['EESC (ppt)']

degree_linear = 1 # for linear polynomial
degree_quadratic = 2 # for quadratic polynomial
degree_cubic = 3 # we want a cubiuc fit for our polyfit, so the degree for this is 3

# calculating value of coefficients in case of linear polynomial
z_mp = np.polyfit(x_mp, y_mp, degree_linear)
print("\ncoefficient value in case of linear polynomial:\n", z_mp)

# calculating value of coefficient in case of quadratic polynomial
z1_mp = np.polyfit(x_mp, y_mp, degree_quadratic)
print("\ncoefficient value in case of quadratic polynomial:\n", z1_mp)

# calculating value of coefficient in case of cubic polynomial
z2_mp = np.polyfit(x_mp, y_mp, degree_cubic)
print("\ncoefficient value in case of cubic polynomial:\n", z2_mp)

# now do the fits for all the other scenarios

z2_la = np.polyfit(x_la, y_la, degree_cubic)
z2_ca = np.polyfit(x_ca, y_ca, degree_cubic)
z2_ba = np.polyfit(x_ba, y_ba, degree_cubic)



# calculating value of coefficient for exponential function
exp_mp = np.polyfit(x_mp, np.log(y_mp), 1)
print("\ncoefficient value in case of exponential function (ln(y) = c + m(x)) - have to e to the power of these:\n", exp_mp)
```

```python
# now construct the equations determining the fits

from matplotlib import pyplot as plt
#plt.style.use('seaborn-darkgrid')
plt.rcParams['figure.figsize'] = (16, 10)

from scipy.stats import sem
import math

from sklearn.metrics import r2_score

import matplotlib.pyplot as plt
from importlib import reload
plt=reload(plt)


a_exp = math.exp(2.83139695e-02)
b_exp = math.exp(-4.89233392e+01)

mp_lin_fit = (9.64344018e+01*df_NoMP1987['Time (year)']) + -1.90655699e+05
mp_quad_fit = (1.42765098e+00*df_NoMP1987['Time (year)']**2) + (-5.65219137e+03*df_NoMP1987['Time (year)']) + 5.59595424e+06
mp_cubic_fit = (z2_mp[0]*df_NoMP1987['Time (year)']**3) + (z2_mp[1]*df_NoMP1987['Time (year)']**2) + (z2_mp[2]*df_NoMP1987['Time (year)']) + (z2_mp[3])
mp_exp_fit = b_exp * a_exp**df_NoMP1987['Time (year)']

la_cubic_fit = (z2_la[0]*df_NoLA1990['Time (year)']**3) + (z2_la[1]*df_NoLA1990['Time (year)']**2) + (z2_la[2]*df_NoLA1990['Time (year)']) + (z2_la[3])
ca_cubic_fit = (z2_ca[0]*df_NoCA1992['Time (year)']**3) + (z2_ca[1]*df_NoCA1992['Time (year)']**2) + (z2_ca[2]*df_NoCA1992['Time (year)']) + (z2_ca[3])
ba_cubic_fit = (z2_ba[0]*df_NoBA1999['Time (year)']**3) + (z2_ba[1]*df_NoBA1999['Time (year)']**2) + (z2_ba[2]*df_NoBA1999['Time (year)']) + (z2_ba[3])

# now perform the standard error of the mean for all of these

data = np.array(mp_cubic_fit)

sem_mp = sem(data)
sem_la = sem(la_cubic_fit)
sem_ca = sem(ca_cubic_fit)
sem_ba = sem(ba_cubic_fit)

# now calculate the R2 score for each of these
# y is the actual value and f is the predicted value

y_mp = df_NoMP1987['EESC (ppt)']
y_la = df_NoLA1990['EESC (ppt)']
y_ca = df_NoCA1992['EESC (ppt)']
y_ba = df_NoBA1999['EESC (ppt)']

f_mp = mp_cubic_fit
f_la = la_cubic_fit
f_ca = ca_cubic_fit
f_ba = ba_cubic_fit

# now calclate the r2 score

r2_mp = r2_score(y_mp,f_mp)
r2_la = r2_score(y_la,f_la)
r2_ca = r2_score(y_ca,f_ca)
r2_ba = r2_score(y_ba,f_ba)



# now create the upper and lower fits as part of dataframes

df_NoMP1987['lower_cubic_fit'] = data - (1.96*sem_mp)
df_NoMP1987['upper_cubic_fit'] = data + (1.96*sem_mp)

df_NoLA1990['lower_cubic_fit'] = la_cubic_fit - (1.96*sem_la)
df_NoLA1990['upper_cubic_fit'] = la_cubic_fit + (1.96*sem_la)

df_NoCA1992['lower_cubic_fit'] = ca_cubic_fit - (1.96*sem_ca)
df_NoCA1992['upper_cubic_fit'] = ca_cubic_fit + (1.96*sem_ca)

df_NoBA1999['lower_cubic_fit'] = ba_cubic_fit - (1.96*sem_ba)
df_NoBA1999['upper_cubic_fit'] = ba_cubic_fit + (1.96*sem_ba)



fig = plt.figure()

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

# NoMP1987

ax1.fill_between(df_NoMP1987.loc[:65,'Time (year)'],df_NoMP1987.loc[:65,'lower_cubic_fit'],df_NoMP1987.loc[:65,'upper_cubic_fit'], color = 'blue', alpha = 0.2)
ax1.scatter(df_NoMP1987['Time (year)'],df_NoMP1987['EESC (ppt)'],color='blue',label='NoMP1987',s=8)
#ax1.plot(df_NoMP1987['Time (year)'],mp_lin_fit,color='red',label='linear fit')
#ax1.plot(df_NoMP1987['Time (year)'],mp_quad_fit,color='blue',label='quadratic fit')
ax1.plot(df_NoMP1987['Time (year)'],data,color='blue')
#ax1.plot(df_NoMP1987['Time (year)'],mp_exp_fit,color='purple',label='exponential fit')
#ax1.xlabel='year'
ax1.text(2019,1200,'R\u00b2 value = {:.4}'.format(r2_mp),color='blue', fontstyle='italic',fontsize=10)
ax1.set_ylabel('EESC (parts per trillion)')
ax1.legend()

# NoLA1990

ax2.scatter(df_NoLA1990['Time (year)'],df_NoLA1990['EESC (ppt)'],color='purple',label='NoLA1990',s=8)
ax2.plot(df_NoLA1990['Time (year)'],la_cubic_fit,color='purple')
ax2.fill_between(df_NoLA1990['Time (year)'],df_NoLA1990['lower_cubic_fit'],df_NoLA1990['upper_cubic_fit'], color = 'purple', alpha = 0.2)
ax2.text(2040,1200,'R\u00b2 value = {:.4}'.format(r2_la),color='purple', fontstyle='italic',fontsize=10)
ax2.legend()

# NoCA1992


ax3.scatter(df_NoCA1992['Time (year)'],df_NoCA1992['EESC (ppt)'],color='orange',label='NoCA1992',s=8)
ax3.plot(df_NoCA1992['Time (year)'],ca_cubic_fit,color='orange')
ax3.fill_between(df_NoCA1992['Time (year)'],df_NoCA1992['lower_cubic_fit'],df_NoCA1992['upper_cubic_fit'], color = 'orange', alpha = 0.2)
ax3.set_ylabel('EESC (parts per trillion)')
ax3.text(2074,1420,'R\u00b2 value = {:.4}'.format(r2_ca),color='orange', fontstyle='italic',fontsize=10)
ax3.legend()

# NoBA1999

ax4.scatter(df_NoBA1999['Time (year)'],df_NoBA1999['EESC (ppt)'],color='green',label='NoBA1999',s=8)
ax4.plot(df_NoBA1999['Time (year)'],ba_cubic_fit,color='green')
ax4.fill_between(df_NoBA1999['Time (year)'],df_NoBA1999['lower_cubic_fit'],df_NoBA1999['upper_cubic_fit'], color = 'green', alpha = 0.2)
ax4.text(1995,740,'R\u00b2 value = {:.4}'.format(r2_ba),color='green', fontstyle='italic',fontsize=10)
ax4.legend()
```
